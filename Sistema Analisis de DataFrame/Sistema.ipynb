```python
# Gráfico de Barras: Suma total de usuarios afectados por cada tipo de ataque
attack_type_affected_users = df.groupby('Attack Type')['Number of Affected Users'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x=attack_type_affected_users.index, y=attack_type_affected_users.values, palette='viridis')
plt.title('Suma Total de Usuarios Afectados por Tipo de Ataque')
plt.xlabel('Tipo de Ataque')
plt.ylabel('Suma de Usuarios Afectados')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Gráfico Circular: Porcentaje del total de usuarios afectados según tipo de ataque.
# Calcular el total global de usuarios afectados
total_global_affected_users = attack_type_affected_users.sum()

# Calcular el porcentaje de usuarios afectados por cada tipo de ataque
attack_type_percentage = (attack_type_affected_users / total_global_affected_users) * 100

plt.figure(figsize=(10, 10))
plt.pie(attack_type_percentage, labels=attack_type_percentage.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(attack_type_percentage)))
plt.title('Porcentaje del Total de Usuarios Afectados según Tipo de Ataque', y=1.02) # y=1.02 ajusta la posición del título
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
# prompt: Diagrama de Pareto: Muestra los tipos de ataque más impactantes y su acumulado porcentual. todo segun el df

# Preparar datos para el Diagrama de Pareto
# Agrupar por tipo de ataque y sumar el número de usuarios afectados
pareto_data = df.groupby('Attack Type')['Number of Affected Users'].sum().sort_values(ascending=False)

# Calcular el total acumulado y el porcentaje acumulado
pareto_data_cumulative = pareto_data.cumsum()
pareto_data_percentage = (pareto_data_cumulative / pareto_data.sum()) * 100

# Crear el Diagrama de Pareto
fig, ax1 = plt.subplots(figsize=(14, 8))

# Gráfico de barras para la suma de usuarios afectados
sns.barplot(x=pareto_data.index, y=pareto_data.values, ax=ax1, palette='viridis')
ax1.set_xlabel('Tipo de Ataque')
ax1.set_ylabel('Suma de Usuarios Afectados', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_title('Diagrama de Pareto de Usuarios Afectados por Tipo de Ataque')
ax1.set_xticklabels(pareto_data.index, rotation=45, ha='right')

# Gráfico de línea para el porcentaje acumulado
ax2 = ax1.twinx()
ax2.plot(pareto_data.index, pareto_data_percentage, color='red', marker='o', linestyle='-')
ax2.set_ylabel('Porcentaje Acumulado', color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.set_ylim(0, 110) # Ajustar el límite superior para ver el 100% claramente

# Añadir etiquetas de porcentaje acumulado en la línea
for i, txt in enumerate(pareto_data_percentage):
    ax2.annotate(f'{txt:.1f}%', (pareto_data.index[i], txt), textcoords="offset points", xytext=(0,10), ha='center')

plt.grid(axis='y', alpha=0.75, linestyle='--')
plt.tight_layout()
plt.show()
```
